```python
# importing the libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
# here N is the total number of stars you want in the system
X_1 = np.random.uniform(0,1,1000) # generating the masses (random masses)
```

```python
X_1
```

```python
radius = 1/np.sqrt(X_1**(-2/3) - 1) # radius of the corresponding masses calculated via the inverse function
radius
```

```python


theta = np.random.uniform(0,1,size=1000)
phi = np.random.uniform(0,1,size=1000)
```

```python
# transformation relations from (r,theta,phi) ----> (x,y,z)

z = (1-2*theta)*radius
x = (radius**2 - z**2)**(1/2)*np.cos(2*np.pi*phi)
y = (radius**2 + z**2)**(1/2)*np.sin(2*np.pi*phi)
```

```python
from mpl_toolkits import mplot3d

fig = plt.figure()

ax = plt.axes(projection = '3d')

ax.plot3D(0,0,0,'gd')
ax.plot3D(x,y,z,'ro')
plt.title('Distrubtion of stars in spatial subspace in plummer halo')
plt.legend(['center of the galaxy','stars in the system'])
plt.show()
```
