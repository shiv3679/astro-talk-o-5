```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_1 = np.random.uniform(0,1,1000) # generating the masses
```

```python
X_1
```

```python
radius = 1/np.sqrt(X_1**(-2/3) - 1) # radius of the corresponding masses
radius
```

```python
# transformation relations from (r,theta,phi) ----> (x,y,z)

theta = np.random.uniform(0,1,size=1000)
phi = np.random.uniform(0,1,size=1000)
```

```python
z = (1-2*theta)*radius
x = (radius**2 - z**2)**(1/2)*np.cos(2*np.pi*phi)
y = (radius**2 + z**2)**(1/2)*np.sin(2*np.pi*phi)
```

```python
from mpl_toolkits import mplot3d

fig = plt.figure()

ax = plt.axes(projection = '3d')

ax.plot3D(0,0,0,'gd')
ax.plot3D(x,y,z,'ro')
plt.show()
```
